# Morphological Operations

```python
# Importing libraries
import cv2
import sys
import imutils
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
np.set_printoptions(threshold=sys.maxsize, linewidth = 120)
```

```python
# Reading a sample image
img = cv2.imread('C://Users//ankushukla//Documents//Work//CV Training//Images//edge.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# img = cv2.GaussianBlur(img, (7,7), 0)
# ret, img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
# img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 115, 1)
# imge = cv2.erode(img, kernel = np.ones((3, 3),np.uint8), iterations = 3)
```

```python
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
# i = cv2.GaussianBlur(img,(5,5),0)
e = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
plt.imshow(e, cmap='gray', vmin=0, vmax=255)
plt.title('Image showing horizontal edges detected by Sobel operator')
plt.show()
```

```python
kernel = np.ones((5,5),np.uint8)
kernel[[0,1,3,4],:] = 0
# kernel[1,:] = 0
print(kernel)
erosion = cv2.erode(e,kernel,iterations = 2)
plt.imshow(erosion, cmap='gray', vmin=0, vmax=255)
```

```python
kernel = np.ones((5,5),np.uint8)
kernel[:, [0,1,3,4]] = 0
# kernel[1,:] = 0
print(kernel)
erosion = cv2.erode(e,kernel,iterations = 2)
plt.imshow(erosion, cmap='gray', vmin=0, vmax=255)
```

```python
# i = cv2.GaussianBlur(img,(5,5),0)
e = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
plt.imshow(e, cmap='gray', vmin=0, vmax=255)
```

```python
kernel = np.ones((5,5),np.uint8)
kernel[:,[0,1,3,4]] = 0
# kernel[1,:] = 0
print(kernel)
erosion = cv2.erode(e,kernel,iterations = 2)
plt.imshow(erosion, cmap='gray', vmin=0, vmax=255)
```

```python
# Reading a sample image
img = cv2.imread('C://Users//ankushukla//Documents//Work//CV Training//Images//blood.jpg')
plt.imshow(img)
plt.title('Microscopic image of blood cells')
plt.show()
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# img = cv2.GaussianBlur(img, (7,7), 0)
# ret, img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 115, 1)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.title('Blood cell image after adaptive thresholding')
plt.show()
```

```python
M = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
i = 3
e = cv2.dilate(img,M,iterations = i)
plt.imshow(e, cmap='gray', vmin=0, vmax=255)
plt.title('Blood cell image after dilation operation {} times'.format(i))
plt.show()
```

```python
M_ = cv2.getStructuringElement(cv2.MORPH_CROSS,(2,2))
i = 2
e = cv2.erode(e,M_,iterations = 3)
plt.imshow(e, cmap='gray', vmin=0, vmax=255)
plt.title('Blood cell image after erosion operation {} times'.format(i))
plt.show()
```

```python
M = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
i = 3
e = cv2.dilate(e,M,iterations = i)
plt.imshow(e, cmap='gray', vmin=0, vmax=255)
plt.title('Blood cell image after dilation operation {} times'.format(i))
plt.show()
```
