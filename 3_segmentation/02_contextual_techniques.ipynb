# Contextual Image Segmentation Techniques

In dicussion so far we have seen segmentation which work on a global parameter (example: threshold) derived from an image. But there is another set of techniques which are more sophisticated than non-contextual techniques. 

This sophistication comes from the context these techniques bring into the approach. These techniques have shown to yield more refined segements as compared to non-contextual techniques. 

However, they also have their fare share of issues. Let's have a look at these techniques. We will not be looking at the implementation of these techniques but we will be discussing them briefly.

## Region Growing and Splitting

Region growing and dissimilarity techniques are based on a uniformity criterion. This criterion applies over a neighborhood of pixels and segments pixels based on the criterion.

### Region Growing

Region growing techniques approach the problem in a bottom up approach. We start with selecting a few pixels and then apply the uniformity criterion to figure out which regions or pixels are same/similar to our seed pixels. 

Based on the criterion we keep on joining the pixels with our seed pixels and create regions. This technique however suffers from unstability and the results change with small changes in the criterion.

Region splitting works opposite to region growing. It first assumes that the whole image is one segment and then go on dividing the segments based on the criterion.
