```python
# Inputs and Labels
data = [
    [0.72,0.82],
    [0.91,-0.69],
    [0.03,0.93],
    [0.12,0.25],
    [0.96,0.47],
    [0.8,-0.75],
    [0.46,0.98],
    [0.66,0.24],
    [0.72,-0.15],
    [0.35,0.01],
    [-0.11,0.1],
    [0.31,-0.96],
    [0.0,-0.26],
    [-0.43,-0.65],
    [0.57,-0.97],
    [-0.72,-0.64],
    [-0.25,-0.43],
    [-0.12,-0.9],
    [-0.58,0.62],
    [-0.77,-0.76],
    [0.55, -0.8]
]

labels = []
for i in range(len(data)):
    if(i<len(data)/2):
        labels.append([1, 0])
    else:
        labels.append([0, 1])
labels = np.array(labels)
```

```python
get_plot(data, labels, title = 'Scatter plot for the dataset')
```

```python
def slp_train(data, labels, weights, bias, threshold):
    miss = 1
    while(miss > 0): 
        miss = 0
        for i in range(len(data)):
            updates = 0

            output = np.sum(np.multiply(np.matrix(data[i]), weights)) + bias

            if(output > threshold):
                y = 1
            else:
                y = -1
            
            true = labels[i].argmax(axis=0)
            if(not true):
                true = true - 1
                
            if y != true:
                miss += 1
    #           learning rule w' = w + eta*error*x | b' = b + eta*error
                weights = weights + np.multiply(true - y, np.array([data[i][0], data[i][1]]))*learning_rate
                bias = bias + learning_rate*(true - y)
                updates = 1

    #       converting weights and bias to decision boundary
            xA = 1
            xB = -1

            if updates != 0:
                yA = (- weights[0] * xA - bias[0]) / weights[1]
                yB = (- weights[0] * xB - bias[0]) / weights[1]
            else:
                xA = - bias[0] / weights[0]
                xB = - bias[0] / weights[0]

                yA = 1
                yB = -1

            if(updates != 0):
    #             Plot the black line, that is, we want to learn the black line as faithfully as possible
    #             ("Best" Decision Boundary)
    #             plt.plot([0.77, -0.55], [-1, 1], color='k', linestyle='-', linewidth=1)
                plt.clf() # Clear figure
                plt.grid(False) # Plot a grid
                plt.xlim(-1,1) # Set x-axis limits
                plt.ylim(-1,1) # Set y-axis limits
                plt.plot([xA, xB], [yA, yB], color = 'g')
                get_plot(data, labels, title = 'Scatter plot of data with classification')
    
    return (weights, bias)
```

#### Setting Network & Training

```python
weights = np.array([0, 0])
bias = np.array([1])
learning_rate = 1
threshold = 0
```

```python
fweights, fbias = slp_train(data, labels, weights, bias, threshold)
```

```python
print('The final weights and biases are ', fweights, ' ', fbias)
```
