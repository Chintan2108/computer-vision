```python
# Importing libraries
import cv2
import sys
import imutils
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
np.set_printoptions(threshold=sys.maxsize, linewidth = 120)
```

```python
# Reading a sample image
img = cv2.imread('C://Users//ankushukla//Documents//Work//CV Training//edge.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img = cv2.GaussianBlur(img, (7,7), 0)
# ret, img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 115, 1)
# imge = cv2.erode(img, kernel = np.ones((3, 3),np.uint8), iterations = 3)
```

```python
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```

```python
e = cv2.Canny(img,20,200)
plt.imshow(e, cmap='gray', vmin=0, vmax=255)
```

```python
# i = cv2.GaussianBlur(img,(5,5),0)
e = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
plt.imshow(e, cmap='gray', vmin=0, vmax=255)
```

```python
e = cv2.Laplacian(img,cv2.CV_64F)
plt.imshow(e, cmap='gray', vmin=0, vmax=255)
```
