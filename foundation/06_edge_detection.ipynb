```python
# Importing libraries
import cv2
import sys
import imutils
import numpy as np
sys.path.append('../')
import matplotlib.pyplot as plt
from basics import functions as fn
import matplotlib.gridspec as gridspec
np.set_printoptions(threshold=sys.maxsize, linewidth = 120)
```

```python
# Reading a sample image
img = cv2.imread('C://Users//ankushukla//Documents//Work//CV Training//Images//taj.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.title('Sudoku Image')
plt.show()
```

```python
sobelx = np.array([[1,2,1], [0,0,0],[-1,-2,-1]])
sobely = np.array([[-1,0,1], [-2,0,2],[-1,0,1]])
laplacian = np.array([[0,-1,0], [-1,4,-1],[0,-1,0]])
print(sobelx)
print(sobely)
print(laplacian)
gradientx = cv2.filter2D(img, -1, sobelx)
gradienty = cv2.filter2D(img, -1, sobely)
lapImg = cv2.filter2D(img, -1, laplacian)

f, (ax1, ax2, ax3, ax4) = plt.subplots(1,4, figsize = (50,50))
ax1.imshow(img, cmap='gray', vmin=0, vmax=255)
ax2.imshow(gradientx, cmap='gray', vmin=0, vmax=255)
ax3.imshow(gradienty, cmap='gray', vmin=0, vmax=255)
ax4.imshow(lapImg, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
e1 = cv2.imread('C://Users//ankushukla//Documents//Work//CV Training//Images//edge1.png')
e1 = cv2.cvtColor(e1, cv2.COLOR_BGR2RGB)
e2 = cv2.imread('C://Users//ankushukla//Documents//Work//CV Training//Images//edge2.png')
e2 = cv2.cvtColor(e2, cv2.COLOR_BGR2RGB)
e3 = cv2.imread('C://Users//ankushukla//Documents//Work//CV Training//Images//edge3.png')
e3 = cv2.cvtColor(e3, cv2.COLOR_BGR2RGB)
e4 = cv2.imread('C://Users//ankushukla//Documents//Work//CV Training//Images//edge4.png')
e4 = cv2.cvtColor(e4, cv2.COLOR_BGR2RGB)
f, (ax1, ax2,ax3,ax4) = plt.subplots(1,4, figsize = (40,10))
ax1.imshow(e1)
ax2.imshow(e2)
ax3.imshow(e3)
ax4.imshow(e4)
plt.show()
```

```python
e1 = e1[:,:,2]
e2 = e2[:,:,2]
e3 = e3[:,:,2]
e4 = e4[:,:,2]
f, (ax1, ax2,ax3,ax4) = plt.subplots(1,4, figsize = (40,10))
ax1.imshow(e1, cmap='gray', vmin=0, vmax=255)
ax2.imshow(e2, cmap='gray', vmin=0, vmax=255)
ax3.imshow(e3, cmap='gray', vmin=0, vmax=255)
ax4.imshow(e4, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
x1 = cv2.filter2D(e1, -1, sobelx)
x2 = cv2.filter2D(e2, -1, sobelx)
x3 = cv2.filter2D(e3, -1, sobelx)
x4 = cv2.filter2D(e4, -1, sobelx)
f, (ax1, ax2,ax3,ax4) = plt.subplots(1,4, figsize = (40,10))
ax1.imshow(x1, cmap='gray', vmin=0, vmax=255)
ax2.imshow(x2, cmap='gray', vmin=0, vmax=255)
ax3.imshow(x3, cmap='gray', vmin=0, vmax=255)
ax4.imshow(x4, cmap='gray', vmin=0, vmax=255)
plt.show()
```
