# Playing With Image Brightness and Contrast

Notebook to look at some examples to understand image brightness and contrast using Python

```python
# Importing libraries
import cv2
import sys
import imutils
import numpy as np
sys.path.append('../')
import matplotlib.pyplot as plt
from basics import functions as fn
from PIL import Image, ImageEnhance 
import matplotlib.gridspec as gridspec
np.set_printoptions(threshold=sys.maxsize, linewidth = 120)
```

```python
im = Image.open("C://Users//ankushukla//Documents//Work//CV Training//Images//cat.jpg")
enhancer = ImageEnhance.Brightness(im)
enhanced_im = enhancer.enhance(1.8)
```

```python
plt.imshow(im, cmap='gray', vmin=0, vmax=255)
plt.title('A grayscale image of a cat')
plt.show()
```

```python
plt.imshow(enhanced_im, cmap='gray', vmin=0, vmax=255)
plt.title('A grayscale image of a cat with enhanced brightness')
plt.show()
```

## Image Contrast

```python
# Reading input image
fog = cv2.imread('C://Users//ankushukla//Documents//Work//CV Training//Images//fog.jpg')
fog = cv2.cvtColor(fog, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fog)
plt.title('The infamous Delhi fog')
plt.show()
```

```python
plt.plot(cv2.calcHist([fog[:,:,0]],[0],None,[256],[0,256]), color = 'red')
plt.plot(cv2.calcHist([fog[:,:,1]],[0],None,[256],[0,256]), color = 'green')
plt.plot(cv2.calcHist([fog[:,:,2]],[0],None,[256],[0,256]), color = 'blue')
plt.show()
```

### Perceptual Contrast

```python
# Creating a simple 2D matrix
box1 = np.full((10,10), 255)
box1[2:8, 2:8] = 150
box1
```

```python
box2 = np.full((10,10), 123)
box2[2:8, 2:8] = 150
box2
```

```python
box3 = np.full((10,10), 61)
box3[2:8, 2:8] = 150
box3
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1,3)
ax1.imshow(box1, cmap='gray', vmin=0, vmax=255)
ax2.imshow(box2, cmap='gray', vmin=0, vmax=255)
ax3.imshow(box3, cmap='gray', vmin=0, vmax=255)
plt.show()
```
