# What do we mean by Image Resolution and how many types of them are there?

This notebook gives an overview of what is the meaning of image resolution of spatial, spectral and temporal kind

```python
# Importing libraries
import cv2
import sys
import imutils
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
np.set_printoptions(threshold=sys.maxsize, linewidth = 120)
```

```python
# Reading a sample image
img = cv2.imread('..//images//jaipur-1.jpg')
image = img.copy()
```

```python
image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
plt.title('Grayscale image of a cat')
plt.show()
```

### Function to Implement Spatial Binning

#### Using Two FOR Loops

```python
# Binning function
def spatial_binning(img, fac = 2):
    res = np.ones(img.shape)
    for i in range(0, res.shape[0], fac):
        for j in range(0, res.shape[1], fac):
            res[i:i+fac, j:j+fac] = np.average(img[i:i+fac, j:j+fac])
    res[:,:] = 255*res[:,:]/np.max(res[:,:])
    return res
```

#### Vectorized Version

```python
res = spatial_binning(image, fac = 64)
f, (ax1, ax2) = plt.subplots(1, 2, figsize = (50,20))
f.suptitle('Lowering the spatial resolution of an image', fontsize=64)
ax1.imshow(image, cmap='gray', vmin=0, vmax=255)
ax1.set_title('Original Cat Image', fontsize=64)
ax2.imshow(res, cmap='gray', vmin=0, vmax=255)
ax2.set_title('Low Resolution Cat Image', fontsize=60)
plt.show()
```
