## Importing Images In Python

Let's see how we can import and visualize image using different Python packages.

```python
# Importing libraries

import sys
import cv2
import numpy as np
from PIL import Image
from skimage import io
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
np.set_printoptions(threshold=sys.maxsize, linewidth = 120)
```

```python
# Reading an image in Python using OpenCV

image = cv2.imread('..//assets//stock//rupee.jpg')
image_ = cv2.resize(image, (20, 30))
print('Image size {}'.format(image.shape))
image_[:, :, 0]
```

```python
# Visualizing the Image using Matplotlib. Image Ref: https://www.vecteezy.com/vector-art/658591-indian-rupee-sign

plt.imshow(image_)
plt.title('Indian Rupee Symbol')
plt.axis('off')
plt.show()
```

Each number in the above matrix is called a **Pixel**. Pixels are the basic unit of an image. The pixel is the level to which we can distinguish or perceive change in the grayscale intensity change in an image.

For a 3 channel RGB image. Each pixel is a 3 dimensional vector, one dimension for each of Red, Green and Blue color.

```python
# Reading a sample image

path = '..//assets//images//goa-1.jpg'

im_cv2 = cv2.imread(path)
im_pil = Image.open(path)
im_skimage = io.imread(path)
im_plt = plt.imread(path)
```

```python
# Displaying the images

f, ax = plt.subplots(2, 2, figsize = (10, 10))
ax[0][0].imshow(im_cv2)
ax[0][1].imshow(im_pil)
ax[1][0].imshow(im_skimage)
ax[1][1].imshow(im_plt)

for x, lib in zip([ax[0][0], ax[0][1], ax[1][0], ax[1][1]], ['OpenCV', 'Pillow', 'Scikit-Image', 'Matplotlib']):
    x.set_title(lib)
    x.axis('off')
    
plt.suptitle('Reading Image in Python Using Different Utitlities. Viewing them using Matplotlib')
plt.show()
```

## How Are Colored Images Stored

Colored images are stored as 3 dimensional matrices. The three dimensions are formed by three 2D marices, each representing Red, Gree and Blue wavelenghts. For satellite images there can be many more such images representing a wider range of the electromagnetic spectrum.

Let's look at these 3 matrices seperately.

```python
f, ax = plt.subplots(1, 4, figsize = (20, 5))
ax[0].imshow(im_skimage)
ax[1].imshow(im_skimage[:, :, 0], cmap = 'gray')
ax[2].imshow(im_skimage[:, :, 1], cmap = 'gray')
ax[3].imshow(im_skimage[:, :, 2], cmap = 'gray')

for x, c in zip(ax, ['Complete', 'Red', 'Green', 'Blue']):
    x.axis('off')
    x.set_title(c)

f.suptitle('Visualizing different color channels seperately')
plt.show()
```

```python
# By default OpenCV reads the image as BGR which needs to be converted to RGB. Let's do that.
# We can also use the same function to create grayscale images

f, ax = plt.subplots(1, 2, figsize = (20, 10))

im_cv2_rgb = cv2.cvtColor(im_cv2, cv2.COLOR_BGR2RGB)
im_cv2_gray = cv2.cvtColor(im_cv2, cv2.COLOR_BGR2GRAY)

ax[0].imshow(im_cv2_rgb)
ax[1].imshow(im_cv2_gray, cmap = 'gray')

for x, c in zip(ax, ['RGB Image', 'Grayscale Image']):
    x.set_title(c)
    x.axis('off')
    
f.suptitle('Conversion of a BGR image to RGB and Grayscale')
plt.show()
```

```python
print('Shape of an RGB Image {}'.format(im_cv2_rgb.shape))
print('Shape of an Grayscale Image {}'.format(im_cv2_gray.shape))
```
